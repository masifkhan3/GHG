```python
# GHG Emission Calculator for Coal and Natural Gas

def calculate_coal_emissions(amount_of_coal):
    """
    Calculate CO2 emissions based on the amount of coal burned.

    Parameters:
    amount_of_coal (float): Amount of coal in kilograms.

    Returns:
    float: CO2 emissions in kilograms.
    """
    emission_factor_coal = 2.2  # kg CO2 per kg of coal
    emissions = amount_of_coal * emission_factor_coal
    return emissions

def calculate_gas_emissions(amount_of_gas):
    """
    Calculate CO2 emissions based on the amount of natural gas burned.

    Parameters:
    amount_of_gas (float): Amount of gas in cubic meters.

    Returns:
    float: CO2 emissions in kilograms.
    """
    emission_factor_gas = 1.8  # kg CO2 per m³ of gas
    emissions = amount_of_gas * emission_factor_gas
    return emissions

# User input for calculations
def main():
    print("Greenhouse Gas Emission Calculator\n")

    # Input for coal
    amount_of_coal = float(input("Enter the amount of coal burned (kg): "))
    coal_emissions = calculate_coal_emissions(amount_of_coal)
    print(f"CO2 emissions from burning {amount_of_coal} kg of coal: {coal_emissions:.2f} kg CO2")

    # Input for natural gas
    amount_of_gas = float(input("Enter the amount of natural gas burned (m³): "))
    gas_emissions = calculate_gas_emissions(amount_of_gas)
    print(f"CO2 emissions from burning {amount_of_gas} m³ of natural gas: {gas_emissions:.2f} kg CO2")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
# GHG Emission Calculator for Coal, Biomass, and Natural Gas

def calculate_coal_emissions(amount_of_coal):
    """
    Calculate CO2, CH4, and N2O emissions based on the amount of coal burned.

    Parameters:
    amount_of_coal (float): Amount of coal in tons.

    Returns:
    tuple: CO2, CH4, N2O emissions in kilograms.
    """
    emission_factor_coal_co2 = 2200  # kg CO2 per ton of coal
    emission_factor_coal_ch4 = 0.03   # kg CH4 per ton of coal
    emission_factor_coal_n2o = 0.001  # kg N2O per ton of coal

    co2_emissions = amount_of_coal * emission_factor_coal_co2
    ch4_emissions = amount_of_coal * emission_factor_coal_ch4
    n2o_emissions = amount_of_coal * emission_factor_coal_n2o
    return co2_emissions, ch4_emissions, n2o_emissions

def calculate_biomass_emissions(amount_of_biomass):
    """
    Calculate CO2, CH4, and N2O emissions based on the amount of biomass burned.

    Parameters:
    amount_of_biomass (float): Amount of biomass in tons.

    Returns:
    tuple: CO2, CH4, N2O emissions in kilograms.
    """
    emission_factor_biomass_co2 = 1800  # kg CO2 per ton of biomass
    emission_factor_biomass_ch4 = 0.05    # kg CH4 per ton of biomass
    emission_factor_biomass_n2o = 0.002   # kg N2O per ton of biomass

    co2_emissions = amount_of_biomass * emission_factor_biomass_co2
    ch4_emissions = amount_of_biomass * emission_factor_biomass_ch4
    n2o_emissions = amount_of_biomass * emission_factor_biomass_n2o
    return co2_emissions, ch4_emissions, n2o_emissions

def calculate_gas_emissions(amount_of_gas):
    """
    Calculate CO2, CH4, and N2O emissions based on the amount of natural gas burned.

    Parameters:
    amount_of_gas (float): Amount of gas in MMBtu.

    Returns:
    tuple: CO2, CH4, N2O emissions in kilograms.
    """
    emission_factor_gas_co2 = 53.06  # kg CO2 per MMBtu of gas
    emission_factor_gas_ch4 = 0.001   # kg CH4 per MMBtu of gas
    emission_factor_gas_n2o = 0.0001  # kg N2O per MMBtu of gas

    co2_emissions = amount_of_gas * emission_factor_gas_co2
    ch4_emissions = amount_of_gas * emission_factor_gas_ch4
    n2o_emissions = amount_of_gas * emission_factor_gas_n2o
    return co2_emissions, ch4_emissions, n2o_emissions

def calculate_total_ghg(co2, ch4, n2o):
    """
    Calculate total GHG emissions in CO2 equivalents.

    Parameters:
    co2 (float): CO2 emissions in kilograms.
    ch4 (float): CH4 emissions in kilograms.
    n2o (float): N2O emissions in kilograms.

    Returns:
    float: Total GHG emissions in kg CO2e.
    """
    ch4_co2e = ch4 * 25  # Convert CH4 to CO2e
    n2o_co2e = n2o * 298  # Convert N2O to CO2e
    total_ghg = co2 + ch4_co2e + n2o_co2e
    return total_ghg

def compare_with_ne_quotient(total_ghg, neq):
    """
    Compare total GHG emissions with National Emission Quotient (NEQ).

    Parameters:
    total_ghg (float): Total GHG emissions in kg CO2e.
    neq (float): National Emission Quotient in kg CO2e.

    Returns:
    str: Comparison result.
    """
    if total_ghg > neq:
        return "Total GHG emissions exceed the NEQ."
    elif total_ghg < neq:
        return "Total GHG emissions are below the NEQ."
    else:
        return "Total GHG emissions are equal to the NEQ."

# User input for calculations
def main():
    print("Greenhouse Gas Emission Calculator\n")

    # Input for coal
    amount_of_coal = float(input("Enter the amount of coal burned (tons): "))
    coal_emissions = calculate_coal_emissions(amount_of_coal)
    print(f"CO2 emissions from burning {amount_of_coal} tons of coal: {coal_emissions[0]:.2f} kg CO2")
    print(f"CH4 emissions from burning {amount_of_coal} tons of coal: {coal_emissions[1]:.4f} kg CH4")
    print(f"N2O emissions from burning {amount_of_coal} tons of coal: {coal_emissions[2]:.4f} kg N2O")

    # Input for biomass
    amount_of_biomass = float(input("Enter the amount of biomass burned (tons): "))
    biomass_emissions = calculate_biomass_emissions(amount_of_biomass)
    print(f"CO2 emissions from burning {amount_of_biomass} tons of biomass: {biomass_emissions[0]:.2f} kg CO2")
    print(f"CH4 emissions from burning {amount_of_biomass} tons of biomass: {biomass_emissions[1]:.4f} kg CH4")
    print(f"N2O emissions from burning {amount_of_biomass} tons of biomass: {biomass_emissions[2]:.4f} kg N2O")

    # Input for natural gas
    amount_of_gas = float(input("Enter the amount of natural gas burned (MMBtu): "))
    gas_emissions = calculate_gas_emissions(amount_of_gas)
    print(f"CO2 emissions from burning {amount_of_gas} MMBtu of natural gas: {gas_emissions[0]:.2f} kg CO2")
    print(f"CH4 emissions from burning {amount_of_gas} MMBtu of natural gas: {gas_emissions[1]:.4f} kg CH4")
    print(f"N2O emissions from burning {amount_of_gas} MMBtu of natural gas: {gas_emissions[2]:.4f} kg N2O")

    # Calculate total GHG emissions
    total_ghg = (coal_emissions[0] + biomass_emissions[0] + gas_emissions[0],
                  coal_emissions[1] + biomass_emissions[1] + gas_emissions[1],
                  coal_emissions[2] + biomass_emissions[2] + gas_emissions[2])

    total_ghg_value = calculate_total_ghg(total_ghg[0], total_ghg[1], total_ghg[2])
    print(f"\nTotal GHG emissions (in kg CO2e): {total_ghg_value:.2f} kg CO2e")

    # Input for NEQ
    neq = float(input("Enter the National Emission Quotient (NEQ) in kg CO2e: "))
    comparison_result = compare_with_ne_quotient(total_ghg_value, neq)
    print(comparison_result)

# Run the main function
if __name__ == "__main__":
    main()
```

```python
def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                raise ValueError("Value must be non-negative.")
            return value
        except ValueError as e:
            print(e)

def calculate_coal_emissions(amount_of_coal):
    emission_factor_coal_co2 = 2200  # kg CO2 per ton of coal
    emission_factor_coal_ch4 = 0.03   # kg CH4 per ton of coal
    emission_factor_coal_n2o = 0.001  # kg N2O per ton of coal

    co2_emissions = amount_of_coal * emission_factor_coal_co2  # in kg
    ch4_emissions = amount_of_coal * emission_factor_coal_ch4  # in kg
    n2o_emissions = amount_of_coal * emission_factor_coal_n2o  # in kg
    return co2_emissions, ch4_emissions, n2o_emissions

def calculate_biomass_emissions(amount_of_biomass):
    emission_factor_biomass_co2 = 1800  # kg CO2 per ton of biomass
    emission_factor_biomass_ch4 = 0.05   # kg CH4 per ton of biomass
    emission_factor_biomass_n2o = 0.002  # kg N2O per ton of biomass

    co2_emissions = amount_of_biomass * emission_factor_biomass_co2  # in kg
    ch4_emissions = amount_of_biomass * emission_factor_biomass_ch4  # in kg
    n2o_emissions = amount_of_biomass * emission_factor_biomass_n2o  # in kg
    return co2_emissions, ch4_emissions, n2o_emissions

def calculate_gas_emissions(amount_of_gas):
    emission_factor_gas_co2 = 53.06  # kg CO2 per MMBtu of gas
    emission_factor_gas_ch4 = 0.001   # kg CH4 per MMBtu of gas
    emission_factor_gas_n2o = 0.0001  # kg N2O per MMBtu of gas

    co2_emissions = amount_of_gas * emission_factor_gas_co2  # in kg
    ch4_emissions = amount_of_gas * emission_factor_gas_ch4  # in kg
    n2o_emissions = amount_of_gas * emission_factor_gas_n2o  # in kg
    return co2_emissions, ch4_emissions, n2o_emissions

# New function for calculating diesel emissions
def calculate_diesel_emissions(amount_of_diesel):
    emission_factor_diesel_co2 = 2.68  # kg CO2 per liter of diesel
    emission_factor_diesel_ch4 = 0.00002  # kg CH4 per liter of diesel
    emission_factor_diesel_n2o = 0.000003  # kg N2O per liter of diesel

    co2_emissions = amount_of_diesel * emission_factor_diesel_co2  # in kg
    ch4_emissions = amount_of_diesel * emission_factor_diesel_ch4  # in kg
    n2o_emissions = amount_of_diesel * emission_factor_diesel_n2o  # in kg
    return co2_emissions, ch4_emissions, n2o_emissions

def calculate_total_ghg(co2, ch4, n2o):
    ch4_co2e = ch4 * 25  # Convert CH4 to CO2e
    n2o_co2e = n2o * 298  # Convert N2O to CO2e
    total_ghg = (co2 + ch4_co2e + n2o_co2e) / 1000  # Convert kg to mton
    return total_ghg

def compare_with_ne_quotient(total_ghg, neq):
    if total_ghg > neq:
        return "Total GHG emissions exceed the NEQ."
    elif total_ghg < neq:
        return "Total GHG emissions are below the NEQ."
    else:
        return "Total GHG emissions are equal to the NEQ."

def main():
    print("Greenhouse Gas Emission Calculator\n")

    # Input for coal
    amount_of_coal = get_positive_float("Enter the amount of coal burned (tons): ")
    coal_emissions = calculate_coal_emissions(amount_of_coal)
    print(f"CO2 emissions from burning {amount_of_coal} tons of coal: {coal_emissions[0]/1000:.2f} mton CO2")
    print(f"CH4 emissions from burning {amount_of_coal} tons of coal: {coal_emissions[1]/1000:.4f} mton CH4")
    print(f"N2O emissions from burning {amount_of_coal} tons of coal: {coal_emissions[2]/1000:.4f} mton N2O")

    # Input for biomass
    amount_of_biomass = get_positive_float("Enter the amount of biomass burned (tons): ")
    biomass_emissions = calculate_biomass_emissions(amount_of_biomass)
    print(f"CO2 emissions from burning {amount_of_biomass} tons of biomass: {biomass_emissions[0]/1000:.2f} mton CO2")
    print(f"CH4 emissions from burning {amount_of_biomass} tons of biomass: {biomass_emissions[1]/1000:.4f} mton CH4")
    print(f"N2O emissions from burning {amount_of_biomass} tons of biomass: {biomass_emissions[2]/1000:.4f} mton N2O")

    # Input for natural gas
    amount_of_gas = get_positive_float("Enter the amount of natural gas burned (MMBtu): ")
    gas_emissions = calculate_gas_emissions(amount_of_gas)
    print(f"CO2 emissions from burning {amount_of_gas} MMBtu of natural gas: {gas_emissions[0]/1000:.2f} mton CO2")
    print(f"CH4 emissions from burning {amount_of_gas} MMBtu of natural gas: {gas_emissions[1]/1000:.4f} mton CH4")
    print(f"N2O emissions from burning {amount_of_gas} MMBtu of natural gas: {gas_emissions[2]/1000:.4f} mton N2O")

    # Input for diesel
    amount_of_diesel = get_positive_float("Enter the amount of diesel oil burned (liters): ")
    diesel_emissions = calculate_diesel_emissions(amount_of_diesel)
    print(f"CO2 emissions from burning {amount_of_diesel} liters of diesel: {diesel_emissions[0]/1000:.2f} mton CO2")
    print(f"CH4 emissions from burning {amount_of_diesel} liters of diesel: {diesel_emissions[1]/1000:.4f} mton CH4")
    print(f"N2O emissions from burning {amount_of_diesel} liters of diesel: {diesel_emissions[2]/1000:.4f} mton N2O")

    # Calculate total GHG emissions
    total_ghg = (
        coal_emissions[0] + biomass_emissions[0] + gas_emissions[0] + diesel_emissions[0],
        coal_emissions[1] + biomass_emissions[1] + gas_emissions[1] + diesel_emissions[1],
        coal_emissions[2] + biomass_emissions[2] + gas_emissions[2] + diesel_emissions[2]
    )

    total_ghg_value = calculate_total_ghg(total_ghg[0], total_ghg[1], total_ghg[2])
    print(f"\nTotal GHG emissions (in mton CO2e) in NIMIR: {total_ghg_value:.2f} mton CO2e")

    # Input for NEQ
    neq = get_positive_float("Enter the National Emission Quotient (NEQ) in mton CO2e in Pakistan: ")
    comparison_result = compare_with_ne_quotient(total_ghg_value, neq)
    print(comparison_result)

# Run the main function
if __name__ == "__main__":
    main()
```
